```python
import pandas as pd
```

In the last notebook, we looked at adding/removing Rows and Columns in Pandas. In this notebook, we explore sorting data in Pandas.

As usual, we start by creating a dictionary in pure Python and then turning it into a Pandas DataFrame to see how to apply changes on a smaller DataFrame, before moving to a bigger real-world example: 

```python
people = {'First name': ['Adam', 'John', 'Jake', 'Jane', 'Job', 'Brock', 'Brian'],
         'Last name': ['Smith', 'Doe', 'Doe', 'Snow', 'Moe', 'Moe', 'Moe'],
         'Email': ['adamsmith@gmail.com', 'johndoe@notreally.com', 'jakedoe@notawebsite.org', 'janesnow@mail.com', 
                   'JobM@mail.com', 'BrockM@mail.com', 'BrianM@mail.com']}
```

```python
people_df = pd.DataFrame(people)
```

```python
people_df
```

Let's say we want to sort our DataFrame.

One way we might do that here is to sort it by last name.

We can do that using the **sort_values** method within Pandas:

```python
people_df.sort_values(by='Last name')
```

As we see, the DataFrame gets sorted alphabetically. If the values within our chosen column were numerical, they would have also been sorted in an ascending order.

If we wish to reverse the order, we can make it a descending order by using the flag **ascending=False**:

```python
people_df.sort_values(by='Last name', ascending=False)
```

**Note: sort_values only views the proposed changes. To make these changes permanent, one needs to use the inplace=True flag.**

Now, when we're dealing with much larger Datasets and projects, it might come in handy to be able to sort by multiple columns at the same time. For example, if multiple people have the same last name, then we want those to be displayed according to the order of a different column, say, their first name. For that effect, we pass a list of columns' names to **sort_values** where the order of columns in that list means precedence in priority.

For instance:

```python
people_df.sort_values(by=['Last name', 'First name'], ascending=False)
```

As we can see above, the last name takes precedence in sorting, and whithin that sorting, duplicates get prioritized based on our chosen sorting of the first name.

Now, for an even more nuanced control over the sorting of the data, we might want to specify different sorting directions (Ascending / Descending) for our different columns, but within the same sorting, for that, we can pass a list of booleans to the **ascending** flag:

```python
people_df.sort_values(by=['Last name', 'First name'], ascending=[False, True])
```

**Take a second to compare the results of the last two cells and undestand the difference using a list of sorting direction creates.**

If at any point we wanted to go back to the original sorting of the DataFrame, i.e. by index, we can do that using the **sort_index** method, which also takes an **inplace** flag:

```python
people_df.sort_index(inplace=True)
```

One more thing to note about sorting in Pandas is that we can sort a single Series without dealing with the entire DataFrame should we wish to, by simply accessing this Series and then calling the **sort_values** method:

```python
people_df['Email'].sort_values()
```

Now, let's see how we might use what we've learned on the Stack Overflow dataset.

First, we load our files, then we change the default options of maximum visible rows and colums to 85 each:

```python
df = pd.read_csv('survey_results_public.csv')
schema_df = pd.read_csv('survey_results_schema.csv')
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

Let's look at the head of our dataset again:

```python
df.head()
```

One obvious way of sorting this dataset might be to sort by country. Let's try that:

```python
df.sort_values(by='Country')
```

A more realistic use of **sort_values** with such a dataset as this one could be to sort by country followed by salary at the same time to see all kinds of information about the income differences within all listed countries. Let's do that, and access the first 50 results:

```python
df.sort_values(by=['Country','ConvertedComp']).head(50)
```

We might want to isolate our 2 target columns to view the results, and if we don't want to change our DataFrame permanently, we can apply the changes to a copy of it:

```python
df2 = df.copy()
```

```python
df2.sort_values(by=['Country', 'ConvertedComp'], ascending=[True, False], inplace=True)
```

```python
df2[['Country', 'ConvertedComp']].head(50)
```

As we can see, we can get a quick view of what people are reporting as their salaries in a given country using a quick sort, and this only gets more useful when combined with the various other powerful capabilities of Pandas.

Now, say we're analyzing a DataFrame in search of the highest results within a certain column within it. We can quickly find these using the **nlargest** method. For instance, to find the highest 20 reported salaries within the survey, we can use:

```python
df2['ConvertedComp'].nlargest(20)
```

A slightely different way of using the **nlargest** method gives us the additional option of **viewing the entire rows which contain these large results**, which is:

```python
df2.nlargest(20, 'ConvertedComp')
```

**Note: There is a companion method to nlargest within Pandas, named nsmallest, which works exactly the same way and gives the samllest results within a Series and can also show the rows containing them, similar to how nlargest does.**
